```python
import matplotlib.pyplot as plt

# Path to your file
filename = "your_data_file.txt"  # Change this to your actual filename

# Read data from file
with open(filename, 'r') as f:
    data = [float(line.strip()) for line in f if line.strip()]

# Create x values (indices)
x = list(range(len(data)))

# Plot
plt.figure(figsize=(8, 5))
plt.plot(x, data, marker='o', linestyle='-')
plt.xlabel('Index')
plt.ylabel('Value')
plt.title('Data Plot')
plt.grid(True)
plt.tight_layout()
plt.show()
```
